```python
# Paso la tabla unificada a SQL para poder visualizarlas mejor y seguir normalizando los datos.
from sqlalchemy import create_engine
engine = create_engine('mysql+pymysql://root:Na3638ty@localhost/peliculas')

df_unificado.to_sql('df_unificado',con=engine,index=False,if_exists='replace')
```

```python
# Cambiar el nombre de las columnas del DF unificado asi podemos identificar rápidamente el contenido.
df_unificado.rename(columns={'type':'Tipo_Film',
                        'title':'Titulo',
                        'director':'Director',
                        'cast':'Reparto',
                        'release_year':'Ano_Rodaje',
                        'duration':'Duracion',
                        'listed_in':'Genero',
                        'plataform':'Plataforma',},
               inplace=True)
```

```python
# Rellenar los datos vacios con el objetivo de optimizar las consultas.
df_unificado['Duracion'].fillna(0, inplace=True)
df_unificado.fillna('Sin Dato', inplace=True)
```

```python
# Se crea una columna con un indice nuevo ya que la columna show_id	tiene datos repetidos
df_unificado['Indice'] = df_unificado.index
df_unificado = df_unificado[['Indice','Tipo_Film','Titulo','Director','Reparto','Ano_Rodaje','Duracion','Genero','Plataforma',]]
```

```python
# Con esto partimos la columna 'Duracion' en dos para separar los datos enteros de los string.
df_dur=df_unificado["Duracion"].str.split(expand=True) #Se crea un dataframe con los datos de la columna "Duracion" separando el número de la unidad
df_unificado['tiempo_duracion']=df_dur[0]
df_unificado['medida_tiempo_duracion']=df_dur[1]
```

```python
df_unificado['tiempo_duracion'] = df_unificado['tiempo_duracion'].astype('float64') # cambie tipo de dato de la columna porque no me dejaba sacar la mean ya que estaba en int
```

```python
# Se calcula el promedios de duración de películas o series por Plataforma
prom_movies = {}
prom_series = {}
for Plataforma in df_unificado['Plataforma'].unique():
    prom_movies[Plataforma] = round(df_unificado[(df_unificado['tiempo_duracion'] !=0 ) & (df_unificado['Tipo_Film']=='Movie')&(df_unificado['Plataforma']==Plataforma)]['tiempo_duracion'].mean()) #Se halla la media de la duración de películas por plataforma
    prom_series[Plataforma] = round(df_unificado[(df_unificado['tiempo_duracion'] !=0 ) & (df_unificado['Tipo_Film']=='TV Show') & (df_unificado['Plataforma']==Plataforma)]['tiempo_duracion'].mean()) #Se halla la media de la duración de series por plataforma
    df_unificado['tiempo_duracion'][df_unificado['tiempo_duracion'].loc[(df_unificado['tiempo_duracion'] == 0) & (df_unificado['Tipo_Film']=='Movie') & (df_unificado['Plataforma']==Plataforma)].index]=prom_movies[Plataforma]
    df_unificado['tiempo_duracion'][df_unificado['tiempo_duracion'].loc[(df_unificado['tiempo_duracion'] == 0) & (df_unificado['Tipo_Film']=='TV Show') & (df_unificado['Plataforma']==Plataforma)].index]=prom_series[Plataforma]
print(prom_movies, prom_series)
```

```python
#Se convierten los actores en listas dentro del dataframe
df_unificado['Reparto_1'] = df_unificado['Reparto'].str.split("," or ":")
df_unificado.drop(columns=['Reparto'],inplace=True)
df_unificado.rename(columns={'Reparto_1':'Reparto'},inplace=True)
```

```python
#Creación de DataFrame para los actores
dic_actores = {'actor':[],
                'año':[],
                'plataforma':[]} #Se crea un diccionario para luego ser convertido en dataframe
for index, actors in enumerate(df_unificado['Reparto']):
    for actor in actors:
        dic_actores['actor'].append(actor) #Se agrega el actor al diccionario
        dic_actores['año'].append(df_unificado['Ano_Rodaje'][index]) #Se agrega el año correspondiente actor al diccionario
        dic_actores['plataforma'].append(df_unificado['Plataforma'][index]) #Se agrega la plataforma correspondiente al actor al diccionario
df_actores = pd.DataFrame(dic_actores)
#df_actors=df_actors.drop_duplicates().reset_index()
```

```python
df_unificado['Genero_1'] = df_unificado["Genero"].str.split(",")
df_unificado.drop(columns=['Genero'],inplace=True)
df_unificado.rename(columns={'Genero_1':'Genero'}, inplace=True)
```

```python
#Creación de DataFrame para los generos
dic_genero = {'genero':[],
                'plataforma':[]} #Se crea un diccionario para luego ser convertido en dataframe
for index, generos in enumerate(df_unificado['Genero']):
    for genero in generos:
        dic_genero['genero'].append(genero) #Se agrega el genero al diccionario
        dic_genero['plataforma'].append(df_unificado['Plataforma'][index])
df_genero = pd.DataFrame(dic_genero)
```

```python
df_actores.to_csv('actores.csv')
df_genero.to_csv('generos.csv')
df_unificado.to_csv('general.csv')
```

```python
#Función 2
def get_count_plataform(plataforma):   
    return{f"Cantidad de películas en {plataforma} es":df_unificado[(df_unificado['Plataforma']==plataforma)&(df_unificado['Tipo_Film']=='movie')]['Tipo_Film'].count().tolist(),
    f"Cantidad de series en {plataforma} es":df_unificado[(df_unificado['Plataforma']==plataforma)&(df_unificado['Tipo_Film']=='tv show')]['Tipo_Film'].count().tolist()}
```

```python
#Función 3
def get_listedin(genero):
    list_genero = []
    for plataforma in df_unificado['Plataforma'].unique():
        list_genero.append(df_genero[(df_genero['genero']==genero)&(df_genero['plataforma']==plataforma)].shape[0])
    
    maximo = max(list_genero)
    if list_genero.index(maximo)==0:
        return {'Amazon': maximo}
    elif list_genero.index(maximo)==1:
        return {'Disney': maximo}
    elif list_genero.index(maximo)==2:
        return {'Hulu': maximo}
    elif list_genero.index('Netflix',maximo)==2:
        return {'Netflix': maximo}
```

```python
#Función 3
def get_listedin(genero):
    genero={(df_genero[(df_genero['genero']==genero)]['plataforma'].value_counts().idxmax()):
    (df_genero[(df_genero['genero']==genero)]['plataforma'].value_counts().max().tolist())}
    return genero
```

```python
#Función 4
def get_actor(plataforma:str, año:int):
    data = df_actores[(df_actores['año']==año)&(df_actores['plataforma']==plataforma)]['actor']
    # return {data.value_counts().max()}
    if data.value_counts().index[0] == 'sin dato':
        return {data.value_counts().index[1]:data.value_counts()[1].tolist()}
    elif data.value_counts().index[0] != 'sin dato':
        return {data.value_counts().index[0]:data.value_counts()[0].tolist()}
```
